# Helper Function

```python
def quick_plot(x_in, df_input, y_scale = 'log',slider=False):

    fig = go.Figure()
    #country_list = ['Germany','India','US','China']
    for each in df_input.columns:
        fig.add_trace(go.Scatter(x=x_in,
                                y=df_input[each],
                                mode = 'markers+lines',
                                opacity = 0.9,
                                line_width = 2,
                                marker_size = 4,
                                name = each))

# defines overall layout
        fig.update_layout(autosize = True,
                          width = 1920,
                          height = 1080,
                          font = dict(
                          family= "PT Sans, monospace",
                          size = 18,
                          color = "#7f7f7f"))                          
                          #xaxis_title='Time',
                          #yaxis_title="Confirmed infected people (Source: John Hopkins, log-scale)"))

        fig.update_yaxes(type=y_scale),
        fig.update_xaxes(tickangle = -45,
                        nticks = 20,
                        tickfont = dict(size=14,color = "#7f7f7f"))
        if slider==True:
            fig.update_layout(xaxis_rangeslider_visible=True)
    fig.show()
#fig.update_layout(xaxis_rangeslider_visible=True)
```

# Exponential Growth

$N(t) = N_0*2^{t/T}$

```python
def doubling_rate(N_0,t,T_d):
    return N_0*np.power(2,t/T_d)
```

```python
max_days = 1000
doubling_rate(100, np.arange(max_days), 1)

norm_slopes = {
    'doubling every day':doubling_rate(100,np.arange(20),1),
    'doubling every 2 days':doubling_rate(100,np.arange(20),2),
    'doubling every 4 days':doubling_rate(100,np.arange(20),4),
    'doubling every 15 days':doubling_rate(100,np.arange(20),15)
}
```

```python
pd_sync_timelines_w_slope = pd.concat([pd.DataFrame(norm_slopes),pd_sync_timelines],axis=1)
```

```python
quick_plot(pd_sync_timelines_w_slope.date,
          pd_sync_timelines_w_slope.iloc[:,0:8],
          y_scale = 'log',
          slider =True)
```

# Doubling Rate - Piecewise Linear Regression

```python
reg = linear_model.LinearRegression(fit_intercept=True)
l_vec = len(df_analyse['Germany'])
x = np.arange(l_vec-1).reshape(-1, 1)
y = np.array(df_analyse['Germany'][1:])
```

```python
reg.fit(x,y)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg.coef_/reg.intercept_
```

```python
def get_rate_via_regression(in_array):
    y = np.array(in_array)
    x = np.arange(-1,2).reshape(-1, 1)
    
    assert len(in_array)==3
    
    reg.fit(x,y)
    intercept = reg.intercept_
    slope = reg.coef_
    return intercept/slope
```

```python
df_analyse['Germany_DR'] = df_analyse['Germany'].rolling(window=3,min_periods=3).apply(get_rate_via_regression)
```

```python
quick_plot(df_analyse.date,df_analyse.iloc[40:,:],y_scale='linear')
```

```python
def doubling_time(in_array):
    y=np.array(in_array)
    return len(y)*np.log(2)/np.log(y[-1]/y[0])
```

```python
df_analyse['Germany_DT'] = df_analyse['Germany'].rolling(window=3,min_periods=3).apply(get_rate_via_regression)
```

```python
quick_plot(df_analyse.date,df_analyse.iloc[40:,[7,8]],y_scale='linear')
```

```python
country_list = df_analyse.columns[1:]
for each in country_list:
    df_analyse[each+'_DR'] = df_analyse[each].rolling(window=3,min_periods=3).apply(get_rate_via_regression)
```

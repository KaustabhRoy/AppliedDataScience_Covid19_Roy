# Data Preparation

```python
raw_data_base = raw_data.drop(['iso_code', 'continent', 'new_cases',
       'new_cases_smoothed', 'total_deaths', 'new_deaths',
       'new_deaths_smoothed', 'total_cases_per_million',
       'new_cases_per_million', 'new_cases_smoothed_per_million',
       'total_deaths_per_million', 'new_deaths_per_million',
       'new_deaths_smoothed_per_million', 'reproduction_rate', 'icu_patients',
       'icu_patients_per_million', 'hosp_patients',
       'hosp_patients_per_million', 'weekly_icu_admissions',
       'weekly_icu_admissions_per_million', 'weekly_hosp_admissions',
       'weekly_hosp_admissions_per_million', 'total_tests', 'new_tests',
       'total_tests_per_thousand', 'new_tests_per_thousand',
       'new_tests_smoothed', 'new_tests_smoothed_per_thousand',
       'positive_rate', 'tests_per_case', 'tests_units', 'total_boosters',
       'new_vaccinations', 'new_vaccinations_smoothed',
       'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred',
       'people_fully_vaccinated_per_hundred', 'total_boosters_per_hundred',
       'new_vaccinations_smoothed_per_million',
       'new_people_vaccinated_smoothed',
       'new_people_vaccinated_smoothed_per_hundred', 'stringency_index',
       'population_density', 'median_age', 'aged_65_older',
       'aged_70_older', 'gdp_per_capita', 'extreme_poverty',
       'cardiovasc_death_rate', 'diabetes_prevalence', 'female_smokers',
       'male_smokers', 'handwashing_facilities', 'hospital_beds_per_thousand',
       'life_expectancy', 'human_development_index',
       'excess_mortality_cumulative_absolute', 'excess_mortality_cumulative',
       'excess_mortality', 'excess_mortality_cumulative_per_million'],axis = 1)
```

```python
raw_data_base.head()
```

```python
raw_data_base = raw_data_base.rename(columns = {'location':'Country'})
#raw_data_base['people_fully_vaccinated'] = raw_data_base['people_fully_vaccinated'].dropna()
raw_data_base
```

```python
raw_data_base['Country']
```

```python
raw_data_india = raw_data_base[raw_data_base['Country']=='India']
raw_data_spain = raw_data_base[raw_data_base['Country']=='Spain']
raw_data_germany = raw_data_base[raw_data_base['Country']=='Germany']
raw_data_usa = raw_data_base[raw_data_base['Country']=='United States']
raw_data_france = raw_data_base[raw_data_base['Country']=='France']
```

```python
absolute_cases = raw_data_base['total_cases']
population = raw_data_base['population']
vaccinated_population = raw_data_base['people_fully_vaccinated']
date = raw_data_base['date']
```

```python
Covid_infectors = absolute_cases/population*100
Vaccination_percentage = vaccinated_population/population*100
```

```python
Covid_infectors
```

```python
Vaccination_percentage
```

```python
raw_data_base['Infected_Population'] = (Covid_infectors)
raw_data_refined = raw_data_base.rename(columns = {'date':'Date'})
```

```python
Date = raw_data_base['date']
Country = raw_data_base['Country']
raw_data_base_vaccinations = pd.DataFrame({'Date':date,'Country':Country,'Vaccinated_Population':(Vaccination_percentage)})
```

```python
raw_data_base_vaccinations_complete = raw_data_base_vaccinations.dropna()
```

```python
country_list = ['Germany','India','Spain']
```

```python
data_india = raw_data_refined['Country']=='India'
data_germany = raw_data_refined['Country']=='Germany'
data_spain = raw_data_refined['Country']=='Spain'
```

```python
country_data_list = data_india|data_germany|data_spain
```

```python
country_infection_data = raw_data_refined[country_data_list]
country_vaccination_data = raw_data_base_vaccinations_complete[country_data_list]
```

# Plot figures for data visualisation

```python
import plotly.express as px
fig = px.line(country_infection_data, x = "Date", y="Infected_Population", color='Country', title = 'Infected population percent over time')
fig.show()
```

```python
fig = px.line(country_vaccination_data, x = "Date", y = "Vaccinated_Population",color = 'Country', title = 'Vaccinated population percent over time')
fig.show()
```
